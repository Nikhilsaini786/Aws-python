```python
#OOPS
class Student:
    def __init__(self,name):
        self.name=name
    def studentName(self):
        print(self.name)
s=Student('Nikhil')
s.studentName()
```

```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def studentName(self):
        print(self.name)
    def studentAge(self):
        print(self.age)
s=Student('Nikhil',23)
s.studentName()
s.studentAge()
```
