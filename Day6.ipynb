```python
class Student:
  def __init__(self,name,roll_no):
    self.name=name
    self.roll_no=roll_no
  def getData(self):
    print(self.name,self.roll_no)

class Address(Student):
  def __init__(self,name,roll_no,city,state):
    self.city=city
    self.state=state
    super().__init__(name,roll_no)
    #extra(repeat)
    #self.name=name
    #self.roll_no=roll_no
  def getAddress(self):
    print(self.city,self.state)
  a=Address('nikhil',123,'jaipur','Rajasthan')
  a.getData()
  a.getAddress()
```
