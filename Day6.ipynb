```python
class BankAccount:

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(amount, "deposited")

    def show_balance(self):
        print("Current Balance =", self.balance)

acc = BankAccount("Nikhil", 1000)

acc.show_balance()
acc.deposit(500)
acc.show_balance()
```
