#List Method

| Method      | Example                      | Result                    |
| ----------- | ---------------------------- | ------------------------- |
| `append()`  | `fruits.append("orange")`    | अंत में जोड़ता है         |
| `insert()`  | `fruits.insert(1, "grapes")` | दिए गए index पर जोड़ता है |
| `remove()`  | `fruits.remove("banana")`    | value हटाता है            |
| `pop()`     | `fruits.pop()`               | आखिरी item हटाता है       |
| `sort()`    | `fruits.sort()`              | list को sort करता है      |
| `reverse()` | `fruits.reverse()`           | list उल्टी करता है        |
| `count()`   | `fruits.count("apple")`      | occurrences गिनता है      |
| `index()`   | `fruits.index("mango")`      | index बताता है            |
| `clear()`   | `fruits.clear()`             | पूरी list खाली करता है    |


```python
fruits = ["apple", "banana", "mango"]
fruits.append("orange")
print(fruits)
```

#Dictionary method

| Method      | Example                             | Result                  |
| ----------- | ----------------------------------- | ----------------------- |
| `keys()`    | `student.keys()`                    | सभी keys देता है        |
| `values()`  | `student.values()`                  | सभी values देता है      |
| `items()`   | `student.items()`                   | key-value pairs देता है |
| `get()`     | `student.get("name")`               | value प्राप्त करता है   |
| `update()`  | `student.update({"city":"Jaipur"})` | data update करता है     |
| `pop()`     | `student.pop("age")`                | key हटाता है            |
| `popitem()` | `student.popitem()`                 | आखिरी pair हटाता है     |
| `clear()`   | `student.clear()`                   | dictionary खाली करता है |
