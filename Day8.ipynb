```python
#flatten
import numpy as np
arr = np.arange(6)
arr1=arr.reshape(2,3)
print(arr1)
#flatten
print(arr1.flatten())
```

```python
#Ravel
import numpy as np
arr = np.arange(6).reshape(2,3)
print(arr)
#ravel
print(arr.ravel())
```

```python
import numpy as np
arr = np.arange(6).reshape(2,3).ravel()
print(arr)
```

```python
#flatten || copy//Return copy of the original array
import numpy as np
arr = np.arange(6)
arr1=arr.reshape(2,3)
print(arr1)
#flatten
arr_new=arr1.flatten()
arr_new[0]=100
print(arr1)
```

```python
#Ravel//Return only reference/view of the original array
import numpy as np
arr = np.arange(6).reshape(2,3)
print(arr)
#ravel
arr_new = arr.ravel()
arr_new[0]=200
print(arr)
```

```python
#Transpose
import numpy as np
arr = np.arange(6).reshape(2,3)
print(arr)
#transpose
print(arr.transpose())
```

```python
#Transpose
import numpy as np
arr = np.arange(6).reshape(2,3)
print(arr)
#transpose
arr1 = arr.T
print(arr1)
```

```python
#Transpose
import numpy as np
arr = np.arange(6).reshape(2,3).transpose()
print(arr)
```
