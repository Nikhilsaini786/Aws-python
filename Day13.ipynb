# Handle With Missing Value

```python
import numpy as np
import pandas as pd
data={
    "Experience":[1,2,3,4,5],
    "Salary":[100,200,300,400,500]
}
df = pd.DataFrame(data)
print(df)
```

```python
import numpy as np
import pandas as pd
data={
    "Experience":[1,np.nan,3,4,5],
    "Salary":[100,200,300,400,np.nan]
}
df = pd.DataFrame(data)
print(df)
#null value filter|find out
print(df.isnull())
#sum of null
print(df.isnull().sum())
```

```python
#dropna |drop complete row
data={
    "Experience":[1,np.nan,3,4,5],
    "Salary":[100,200,300,400,np.nan],
    "name":["hello","dheraj",'Nikhil','Raj','vinay']

}
df = pd.DataFrame(data)
print(df)
df.dropna(inplace=True)
print(df)
```

```python
#fillna | null -> fill
data={
    "Experience":[1,np.nan,3,4,5],
    "Salary":[100,200,300,400,np.nan],
    "name":["hello","dheraj",'Nikhil','Raj','vinay']

}
df = pd.DataFrame(data)
print(df)
df.fillna(5)
```

```python
import matplotlib.pyplot as plt

# Line Chart
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.subplot(1,2,2)
plt.title("Line Chart")
plt.show()

# Bar Chart
x = ["Apple", "Banana", "Mango", "Orange"]
y = [10, 20, 15, 25]
plt.subplot(1,2,2)
plt.bar(x, y)
plt.title("Bar Chart")
plt.show()

#  Pie Chart
labels = ["Apple", "Banana", "Mango", "Orange"]
sizes = [30, 20, 25, 25]
plt.subplot(1,2,2)
plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()

# Multiple Line Chart
x = [1, 2, 3, 4, 5]

salary = [100, 200, 300, 400, 500]
bonus = [50, 100, 150, 200, 250]
plt.subplot(1,2,2)
plt.plot(x, salary, label="Salary")
plt.plot(x, bonus, label="Bonus")


plt.title("Multiple Line Chart")
plt.tight_layout()



plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Line Chart
plt.subplot(2, 2, 1)
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]
plt.plot(x, y)
plt.title("Line Chart")

# 2. Bar Chart
plt.subplot(2, 2, 2)
x = ["Apple", "Banana", "Mango", "Orange"]
y = [10, 20, 15, 25]
plt.bar(x, y)
plt.title("Bar Chart")

# 3. Pie Chart
plt.subplot(2, 2, 3)
labels = ["Apple", "Banana", "Mango", "Orange"]
sizes = [30, 20, 25, 25]
plt.pie(sizes, labels=labels)
plt.title("Pie Chart")

# 4. Multiple Line Chart
plt.subplot(2, 2, 4)
x = [1, 2, 3, 4, 5]
salary = [100, 200, 300, 400, 500]
bonus = [50, 100, 150, 200, 250]

plt.plot(x, salary, label="Salary")
plt.plot(x, bonus, label="Bonus")
plt.title("Multiple Line")


plt.tight_layout()
plt.legend()
plt.show()
```
