**Conditional** **Statement**

```python
#if
num1 = 10
if num1>2:
  print(num1)
```

```python
num1 = 10
if num1>2:
  print("num1=",num1)
print("if end")
```

```python
num1 = int(input("Enter a number: "))

if num1 > 2:
    print("num1 =", num1)

print("if end")
```

```python
#if elif else
num1 = int(input("Enter a Number"))
num2 = 20
num3 = 30
if num1>num2:
  print("if part run")
elif num1>num3:
  print("elif part run")
else:
  print("else part run")
```

```python
#if elif else
num1 = int(input("Enter a Number"))
num2 = 20
if num1>num2:
  print("if part run")
else:
  print("else part run")
```

```python
marks = int(input("Enter a Marks:"))
if marks>=90:
  print("Grade A")
elif marks>=80:
  print("Grade B")
elif marks>=70:
  print("Grade C")
elif marks>=60:
  print("Grade D")
else:
  print("Grade F")
```

#List

```python
#list
l=[10,20,30]
print(l[0])
print(l[1])
print(l[2])
#negative index
print(l[-1])
print(l[-2])
print(l[-3])
```

```python
#mutable
l[0]=100
l[-1]=200
print(l)
```

#dictionary

```python
dict1 = {"Name": "Nikhil"}
print(dict1)
print(dict1["Name"])
#example
dict1 = {"Full Name": "Nikhil Saini"}
print(dict1)
print(dict1["Full Name"])
```

```python
student = {
    "name": "Nikhil",
    "age": 20,
    "city": "Jaipur"
}
print(student)
print(student["name"])
```

```python
d={"age":20,"name":"Nikhil","city":"Jaipur"}
print(d["age"])
print(d["name"])
print(d["city"])
```
