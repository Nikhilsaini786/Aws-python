**Opertor**
1. arthmetic-> +,-,*,/,//,%,**
2. relational -> <,>=,<=,==,!=
3. Assignment -> =,+=,-=	,*=	,/=
4. logical -> or,not,and
5. Bitwise -> &,`,^,~,<<,	>>



```python
#arthemtic
num1=10
num2=20
print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)
print(num1//num2)
print(num1%num2)
print(num1**num2)
```

**Relational Opertor**

```python
a = 10
b = 20
print(a<b)
print(a>b)
print(a<=b)
print(a>=b)
print(a==b)
print(a!=b)
```

**Assignment Opertor**

```python
x=10
print(x)
x+=5
print(x)
x-=2
print(x)
x*=3
print(x)
x/=2
print(x)
```

**Logical Opertor**

```python
#Number
num1=10
num2=20
print(num1 and num2)
```

```python
#Positive number all = True
#Negative Number all = True
#Zero = False
```

```python
num1=10
num2=20
print(num1>num2 and num1==num2)
print(num1>num2 or num1==num2)
print(not num1>num2)
```
