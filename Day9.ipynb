1. p.sum(a) -> Sum of all elements
2. np.mean(a) -> Average = Sum of elements / Number of elements
3. np.median(a) -> Middle value after sorting
4. np.min(a) -> Smallest value in array
5. np.max(a) -> Largest value in array
6. np.var(a) mean,difference,square,average | (sum = ddof) -> divide by array length
7. np.std(a) -> Standard Deviation = √Variance -> Measures spread of data
8. np.prod(a) -> Multiplication of all elements
9. np.cumsum(a) -> Cumulative (running) sum
10. np.cumprod(a) -> Cumulative (running) product
12. np.argmin(a) -> Index position of minimum value
13. np.argmax(a) -> Index position of maximum value
14. np.abs(a) -> Converts negative values to positive -> Absolute value (distance from zero)
15. np.unique(a) -> Returns unique values only -> Removes duplicates
16. np.percentile(a, 50) -> Finds percentage-based value -> 50th percentile = Median
17. np.quantile(a, 0.5) -> Similar to percentile -> 0.5 = 50%
18. np.ptp(a) -> Range = Max - Min
19. np.any(a) -> True if at least one value is True





```python
#var
import numpy as np
arr = np.array([1,2,3,4,5,6,7])
print(np.var(arr))
#standard derivation
print(np.std(arr))
print(np.prod(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
print(np.argmax(arr))
print(np.argmin(arr))
print(np.abs(arr))
print(np.unique(arr))
print(np.quantile(arr,0.50))
print(np.percentile(arr,50))
print(np.ptp(arr))
print(np.any(arr))
```

```python
import numpy as np

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

# 1. Print complete dataset
print("1. Complete Dataset:")
print(sales_data)

# 2. Shape of array
print("\n2. Shape of Array:")
print(sales_data.shape)

# 3. Number of dimensions
print("\n3. Number of Dimensions:")
print(sales_data.ndim)

# 4. Total number of elements
print("\n4. Total Number of Elements:")
print(sales_data.size)

# 5. Restaurant IDs
print("\n5. Restaurant IDs:")
print(sales_data[:, 0])

# 6. 2021 sales data
print("\n6. 2021 Sales Data:")
print(sales_data[:, 1])

# 7. 2024 sales data
print("\n7. 2024 Sales Data:")
print(sales_data[:, 4])

# 8. Maximum sales value
print("\n8. Maximum Sales Value:")
print(np.max(sales_data[:, 1:]))

# 9. Minimum sales value
print("\n9. Minimum Sales Value:")
print(np.min(sales_data[:, 1:]))

# 10. Total sales of all restaurants in 2021
print("\n10. Total Sales in 2021:")
print(np.sum(sales_data[:, 1]))

# 11. Average sales in 2024
print("\n11. Average Sales in 2024:")
print(np.mean(sales_data[:, 4]))

# 12. Total sales of each restaurant
print("\n12. Total Sales of Each Restaurant:")
print(np.sum(sales_data[:, 1:], axis=1))

# 13. Average sales of each restaurant
print("\n13. Average Sales of Each Restaurant:")
print(np.mean(sales_data[:, 1:], axis=1))
```
